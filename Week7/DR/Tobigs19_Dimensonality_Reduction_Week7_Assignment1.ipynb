### 1-2) ë‹¤ìŒìœ¼ë¡œ ê³µë¶„ì‚° í–‰ë ¬ì„ êµ¬í•©ë‹ˆë‹¤

```python
cov_matrix = np.cov(X_scaled.T)
print(cov_matrix)
```

### 1-3) ì´ì œ ê³ ìœ ê°’ê³¼ ê³ ìœ ë²¡í„°ë¥¼ êµ¬í•©ë‹ˆë‹¤

```python
import numpy.linalg as lin

eigenvalues, eigenvectors = lin.eig(cov_matrix)

print("Eigenvalues: ", eigenvalues)
print("Eigenvectors: ", eigenvectors)
```

```python
eigen_values_matrix = np.diag(eigenvalues) # eigenvalue ê°’ìœ¼ë¡œ ëŒ€ê°í–‰ë ¬ ë§Œë“¤ê¸°ê¸°
reconstructed_cov_matrix = np.dot(eigenvectors, np.dot(eigen_values_matrix, eigenvectors.T))

print("Reconstructed covariance matrix: \n", reconstructed_cov_matrix)
```

### 1-5) ë§ˆì§€ë§‰ìœ¼ë¡œ ê³ ìœ  ë²¡í„° ì¶•ìœ¼ë¡œ ê°’ì„ ë³€í™˜í•©ë‹ˆë‹¤


https://docs.scipy.org/doc/numpy/reference/generated/numpy.concatenate.html

```python
def new_coordinates(X,eigenvectors):
    for i in range(eigenvectors.shape[0]):
        if i == 0:
            new = [X.dot(eigenvectors.T[i])]
        else:
            new = np.concatenate((new,[X.dot(eigenvectors.T[i])]),axis=0)
    return new.T
```

```python
from sklearn.preprocessing import StandardScaler

def MYPCA(X,number):
    # Scaling
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Covariance
    cov_matrix = np.cov(X_scaled.T)
    # eigenvalues, eigenvectors
    eigenvalues, eigenvectors = lin.eig(cov_matrix)
    #np.dot(np.dot(eigenvectors,mat),eigenvectors.T)

    # ê³ ìœ ê°’ ì •ë ¬ ë° ê³ ìœ ê°’ í¬ê¸°ì— ë”°ë¼ ê³ ìœ ë²¡í„°ë„ ì •ë ¬

    def new_coordinates(X,eigenvectors):
        for i in range(eigenvectors.shape[0]):
            if i == 0:
                new = [X.dot(eigenvectors.T[i])]
            else:
                new = np.concatenate((new,[X.dot(eigenvectors.T[i])]),axis=0)
        return new.T

    new_coordinate = new_coordinates(X_scaled, eigenvectors) # ê° ë³€ìˆ˜ë“¤ì— ëŒ€í•´ ê³ ìœ  ë²¡í„° ì¶•ìœ¼ë¡œ ë³€í™˜ ì™„ë£Œ
    idx = list(eigenvalues.argsort()[::-1]) # ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ê³ ìœ ê°’ì˜ index ì €ì¥

    for i in range(number):  # ê³ ìœ ê°’ì´ í° indexë¶€í„° numberë§Œí¼ ê°€ì ¸ì™€ ë³€ìˆ˜ë¡œ ì‚¬ìš©
        if i==0:
            new = [new_coordinate[:, idx.index(i)]]
        else:
            new = np.concatenate(([new, [new_coordinate[:, idx.index(i)]]]), axis=0)


    return new.T #ìƒˆë¡œìš´ ì¶•ìœ¼ë¡œ ë³€í™˜ë˜ì–´ ë‚˜íƒ€ë‚œ ë°ì´í„° return
```

```python
MYPCA(X,3)

# ìƒˆë¡œìš´ ì¶•ìœ¼ë¡œ ì˜ ë³€í™˜ë˜ì–´ì„œ ë‚˜íƒ€ë‚˜ë‚˜ìš”?
# ìœ„ì—ì„œ í–ˆë˜ PCAë‘ì€ ì°¨ì´ê°€ ìˆì„ ìˆ˜ ìˆì–´ìš” 
# ì™œëƒí•˜ë©´ ìœ„ì—ì„œëŠ” ê³ ìœ ê°’ì´ í° ì¶• ìˆœì„œë¡œ ì •ë ¬ì„ ì•ˆí–ˆì—ˆê±°ë“ ìš”
```
