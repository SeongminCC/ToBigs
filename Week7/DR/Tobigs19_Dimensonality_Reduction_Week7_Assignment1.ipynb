### 1-2) 다음으로 공분산 행렬을 구합니다

https://docs.scipy.org/doc/numpy/reference/generated/numpy.cov.html 를 참고하시면 도움이 될거에요

```python
cov_matrix = np.cov(X_scaled.T)
print(cov_matrix)
```

### 1-3) 이제 고유값과 고유벡터를 구합니다

실습코드를 참고해주세요

```python
import numpy.linalg as lin

eigenvalues, eigenvectors = lin.eig(cov_matrix)

print("Eigenvalues: ", eigenvalues)
print("Eigenvectors: ", eigenvectors)
```

### 1-4) 고유값 분해: 고유값 분해의 곱으로 원래 공분산 행렬을 구해봅니다

https://docs.scipy.org/doc/numpy/reference/generated/numpy.dot.html 를 참고해주세요

A = 𝑃Λ𝑃`
이 식 기억나죠?

```python
eigen_values_matrix = np.diag(eigenvalues) # eigenvalue 값으로 대각행렬 만들기기
reconstructed_cov_matrix = np.dot(eigenvectors, np.dot(eigen_values_matrix, eigenvectors.T))

print("Reconstructed covariance matrix: \n", reconstructed_cov_matrix)
```

### 1-5) 마지막으로 고유 벡터 축으로 값을 변환합니다


https://docs.scipy.org/doc/numpy/reference/generated/numpy.concatenate.html

```python
def new_coordinates(X,eigenvectors):
    for i in range(eigenvectors.shape[0]):
        if i == 0:
            new = [X.dot(eigenvectors.T[i])]
        else:
            new = np.concatenate((new,[X.dot(eigenvectors.T[i])]),axis=0)
    return new.T
```

```python
from sklearn.preprocessing import StandardScaler

def MYPCA(X,number):
    # Scaling
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Covariance
    cov_matrix = np.cov(X_scaled.T)
    # eigenvalues, eigenvectors
    eigenvalues, eigenvectors = lin.eig(cov_matrix)
    #np.dot(np.dot(eigenvectors,mat),eigenvectors.T)

    # 고유값 정렬 및 고유값 크기에 따라 고유벡터도 정렬

    def new_coordinates(X,eigenvectors):
        for i in range(eigenvectors.shape[0]):
            if i == 0:
                new = [X.dot(eigenvectors.T[i])]
            else:
                new = np.concatenate((new,[X.dot(eigenvectors.T[i])]),axis=0)
        return new.T

    new_coordinate = new_coordinates(X_scaled, eigenvectors) # 각 변수들에 대해 고유 벡터 축으로 변환 완료
    idx = list(eigenvalues.argsort()[::-1]) # 내림차순으로 고유값의 index 저장

    for i in range(number):  # 고유값이 큰 index부터 number만큼 가져와 변수로 사용
        if i==0:
            new = [new_coordinate[:, idx.index(i)]]
        else:
            new = np.concatenate(([new, [new_coordinate[:, idx.index(i)]]]), axis=0)


    return new.T #새로운 축으로 변환되어 나타난 데이터 return
```

```python
MYPCA(X,3)

# 새로운 축으로 잘 변환되어서 나타나나요?
# 위에서 했던 PCA랑은 차이가 있을 수 있어요 
# 왜냐하면 위에서는 고유값이 큰 축 순서로 정렬을 안했었거든요
```
